```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.legend import Legend


def loadList(fileName):
    with open(fileName, newline='') as csv_file:
        reader = csv.reader(csv_file)
        dataList = list(reader)
    return dataList



col = {"date":0, "open":1}
myData = loadList('TSLAnew.csv')


finalList = []
x = 30
#below is line of code to have graph to only display one point for every month instead of one point for every day of the year (just skips by 21 which is on average the amoung of weekdays/trading days in the year)
#for i in range(1641, len(myData), 21):
for i in range(1641, len(myData)):
    X = myData[i][col["date"]]
    days = X.split('-')
    daysOyear =  (int(days[1]) - 1) * 30 + int(days[2])
    
    
    finalList.append([int(days[0]) + daysOyear/360, float(myData[i][col["open"]]), x])
    x += 30
    
finalList.sort(key=lambda x: x[2])
for i in finalList:
    print(i)
```

```python
df = pd.DataFrame(finalList, columns =['Date','Open', 'L'])
df
```

```python
plt.figure(figsize=(25,15))
x = df.Date
y = df.Open

#this block of code was assisted from online website for help in stylistic choices
plt.style.use("seaborn-dark")
for param in ['figure.facecolor', 'axes.facecolor', 'savefig.facecolor']:
    plt.rcParams[param] = '#0B0B61'  # dark blue
for param in ['text.color', 'axes.labelcolor', 'xtick.color', 'ytick.color']:
    plt.rcParams[param] = '0.9'  # grey


plt.xlabel("Date",fontsize=41,fontweight='bold')
plt.ylabel("Price",fontsize=38,fontweight='bold')

plt.title("Tsla Stock over the past 5 years",fontsize=36,fontweight='bold')
plt.xticks(np.arange(2017, 2022, 1),fontsize=38)
plt.yticks(np.arange(0, 950, 100),fontsize=29)


lines = plt.plot(x, y, color='#58ACFA', linewidth=6)
plt.legend(lines[:1], ['Tsla Stock: +1751%'], fontsize=30, loc='upper left', framealpha=1, frameon=True)
plt.grid(color='#FFFFFF')
plt.show()
```

```python
col = {"Date":0, "Carssold":1}
myData = loadList('Tslacarssold - Sheet1 copy.csv')


finalList = []
x = 30

for i in range(1, len(myData)):
    temp = myData[i][0]
    carsSold = int(myData[i][1].replace(',',''))
    #print(carsSold)
    year, day = temp.split('-')
    year = int(year)
    day = int(day)
    day -= 1
    day *= 30
    timeOyear = (year * 365 + day) / 365
    #print(timeOyear)
    finalList.append([timeOyear, int(myData[i][col["Carssold"]].replace(',','')), x])
    #days = X[0].split('-')
    #print(days)

    x += 30
    
finalList.sort(key=lambda k: k[2])
for i in finalList:
    print(i)
```

```python
df = pd.DataFrame(finalList, columns =["Date","Carssold", "L"])
df
```

```python
plt.figure(figsize=(25,15))
x = df.Date
y = df.Carssold

#this block of code was assisted from online website for help in stylistic choices
plt.style.use("seaborn-dark")
for param in ['figure.facecolor', 'axes.facecolor', 'savefig.facecolor']:
    plt.rcParams[param] = '#0B0B61'  # dark blue
for param in ['text.color', 'axes.labelcolor', 'xtick.color', 'ytick.color']:
    plt.rcParams[param] = '0.9'  # grey
    

plt.xlabel("Date",fontsize=41,fontweight='bold')
plt.ylabel("Cars Sold",fontsize=38,fontweight='bold')

plt.title("Number of Cars Tesla has sold each month in the past 5 years",fontsize=36,fontweight='bold')
#plt.xticks(np.arange(2017, 2022, 1),fontsize=38)
plt.xticks(fontsize=38)
plt.yticks(np.arange(0, 51000, 5000),fontsize=29)
#plt.yticks(fontsize=29)

dot = plt.plot(x, y, color='#58ACFA', linewidth=6)
plt.legend(dot[:1], ['Cars Sold by Tesla: +913%'], fontsize=30, loc='upper left', framealpha=1, frameon=True)
plt.grid()
plt.show()
```
